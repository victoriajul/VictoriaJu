### 7. Filter the Data

```python
df.corr(numeric_only=True)
```

```python
df.isnull().sum()
```

```python
df['gross_total']=df['gross_total'].fillna(df['gross_total'].mode()[0])
df.tail(5)
```

```python
df.isnull().sum()
```

```python
def yearby_decade(year): 
    if year>= 2020: 
        return '2020s'
    elif year>= 2010: 
        return '2010s' 
    elif year>= 2000: 
        return '2000s'
    elif year>= 1990:
        return '1990s'
    elif year>= 1980: 
        return '1980s'
    elif year>= 1970: 
        return '1970s'
    else: 
        return '1960s or before'
```

```python
df['yearby_decade']=df['year'].apply(yearby_decade)
df
```
