### Detecting and resolving missing or invalid values

```python
df.isnull().sum()
```

```python
df_temp = df.dropna().copy()
df_temp
```

### Removing Duplicate records

```python
df_temp.duplicated().sum()
```

### Stardardizing Column names

```python
df_temp.rename(columns={'GVT_TYPE_OF_APPT':'APPT_TYPE'}, inplace=True)
```

```python
df_temp['JOBTITLE'] = df_temp['JOBTITLE'].str.strip()
df_temp['JOBTITLE'] = df_temp['JOBTITLE'].str.title()
```

### Encoding or converting cateorical variabes

```python
df_temp['GRADE_NUM'] = df_temp['GRADE'].astype(str).str.replace(r'\D+', '', regex=True)
df_temp['GRADE_NUM'] = pd.to_numeric(df_temp['GRADE_NUM'], errors='coerce').astype('Int64')
```

### Parsing date strings to date time objects

```python
df_temp['HIREDATE'] = pd.to_datetime(df_temp['HIREDATE_STRING'], errors='coerce')
print(df_temp['HIREDATE'].dtype)
```

```python
df_temp['YEAR'] = df_temp['HIREDATE'].dt.year
```

```python
df_temp['YEAR'] = df_temp['YEAR'].astype('Int64')
```

## Dropping irrelevant and redundant columns

```python
df_temp2 = df_temp.drop(columns=['GRADE']).copy()
df_temp2
```

```python
# OBJECTID already excluded when df_temp2 was created above (drop GRADE, keep rest)
# No additional drop needed here
```

### Standardizing columns to make it easier to create readable visualizations 

```python
main_types = ['CS - Reg', 'CS - Temp', 'MSS - Reg', 'MSS - Temp']
df_temp3 = df_temp2[df_temp2['APPT_TYPE'].isin(main_types)].copy()
```

```python
top_grades = df_temp3['GRADE_NUM'].value_counts().nlargest(10).index
df_temp4 = df_temp3[df_temp3['GRADE_NUM'].isin(top_grades)]
```

```python
df_filtered = df_temp4.copy()
```

### Temporal Trends and Seasonality detection

```python
monthly_avg = df_filtered.groupby(df_filtered['HIREDATE'].dt.to_period('M'))['COMPRATE'].mean()
print(monthly_avg.head(10))

df_filtered['COMPRATE_ROLLING'] = df_filtered['COMPRATE'].rolling(window=7).mean()
```

### Outlier detection and anomalous pattern identification

```python
iso_forest = IsolationForest(contamination=0.05)
iso_forest.fit(df_filtered[['GRADE_NUM', 'COMPRATE']].dropna().values)
df_filtered = df_filtered.copy()
df_filtered['Anomaly_Score'] = iso_forest.predict(
    df_filtered[['GRADE_NUM', 'COMPRATE']].fillna(df_filtered[['GRADE_NUM', 'COMPRATE']].median()).values
)
```
